```python
# Sort the data
df.sort_values(by='life_expectancy')
```

```python
# By default sort values is ascending but we want the numbers to be descending
df.sort_values(by='life_expectancy', ascending = False)
```

```python
# What if we want a top-10? Just add .head(10)
df.sort_values(by='life_expectancy', ascending = False).head(10)
```

```python
# To get the last 2 out of the first 10 we just add tail(2)
df.sort_values(by='life_expectancy', ascending = False).head(10).tail(2)
```

```python
# To get a slice of the dataframe e.g. the things with the index of 90 to 102
df.loc[90:102]
```

```python
# If we only want Asian countries
df.continent == "Asia"
```

```python
# Feed the True/False series 
# Gives every row where something is true
df[df.continent == "Asia"]
```

```python
# Same but instead we find all the countries with a life expectancy above 67
df[df.life_expectancy > 67]
```

```python
# We find the ten largest countries in Africa (by population)
df[df.continent == "Africa"] \
    .sort_values(by='population', ascending = False).head(10)
```

```python
# NO ONE on the internet uses this!!!
df.query("continent == 'Europe'")
```

```python
# We try to calculate gdp per capita
df.gdp / df.population
```

```python
# To greate gpd per capita as a new variable, do this:
df['gpd_per_capita'] = df.gdp / df.population

# And don't forget the []
```

```python
df.head()
```

```python
# We try to find the median gpd per capita
df.gpd_per_capita.median()
```

```python
# And the same for the mean gdp per capita
df.gpd_per_capita.mean()
```

```python
# Step 1. find the average gpd per capita in asia
asian_mean = df[df.continent == "Asia"].gpd_per_capita.mean()
asian_mean
```

```python
# And then we find the countries where the continent is Africa AND where the 
# gpd per capita is lower than the mean gpd per capita in Asia
df[(df.continent == 'Africa') & (df.gpd_per_capita < asian_mean)]
```

```python
# Mean life expectancy
df.life_expectancy.mean()
```

```python
# Median life expectancy 
df.life_expectancy.median()
```

```python
# If I want to do a bunch of stuff at once
# .describe() summarizes the data
df.life_expectancy.describe()
```

```python
# Count countries in each continent
df.continent.value_counts()
# If you want to count things, don't use groupby. Use value.counts !!!!!!!!!!
```

```python
# We use groupby to get the median life expectancy for each continent.
df.groupby('continent').life_expectancy.median()
```

```python
# Total population per continent.
df.groupby('continent').population.sum()
```

```python
# Find the percentage
df.continent.value_counts(normalize=True)
```

```python
# Find percentages that aren't decimals
df.continent.value_counts(normalize=True) * 100
```

```python
# Create a new column with population in millions
df['pop_mio'] = df.population / 1000000
df.head()
```
