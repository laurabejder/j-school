### Hm, `%` isn't part of  a number. Let's remove it.

When you're confident you got it right, save the results back into the `ABV` column.

- *Tip: In programming the easiest way to remove something is to *replacing it with nothing*.
- *Tip: "nothing" might seem like `NaN` sinc we talked about it a lot in class, but in this case it isn't! It's just an empty string, like ""*

Save the results back into the `ABV` column (again), and then check `df.dtypes` to make sure it worked.

````python
#And then we convert ABV into a float (to preserve the decimals)
df.ABV = df.ABV.astype(float)

df.dtypes
````

# Measuring International Bitterness Units

## Display all of the IPAs

Include American IPAs, Imperial IPAs, and anything else with "IPA in it."

IPA stands for [India Pale Ale](https://www.bonappetit.com/story/ipa-beer-styles), and is probably the most popular kind of beer in the US for people who are drinking [craft beer](https://www.craftbeer.com/beer/what-is-craft-beer).

````python
df[df.Beer.str.contains('IPA', na=False)]
````

IPAs are usually pretty hoppy and bitter. IBU stands for [International Bitterness Unit](http://www.thebrewenthusiast.com/ibus/), and while a lot of places like to brag about having the most bitter beer (it's an American thing!), IBUs don't necessary *mean anything*.

Let's look at how different beers have different IBU measurements.

## Plot those two histograms on top of one another

To plot two plots on top of one another, you do two steps.

1. First, you make a plot using `plot` or `hist`, and you save it into a variable called `ax`.
2. You draw your second graph using `plot` or `hist`, and send `ax=ax` to it as a parameter.

It would look something like this:

```python
ax = df.plot(....)
df.plot(ax=ax, ....)
``` 

(...except totally different)

````python
# First I save the IPA histogram as ax
ax = df[df.Style.str.contains('IPA.*', na=False)].IBUs.hist(bins=25)

# And then I include it in the wheat beer histogram
df[wheat_beers].IBUs.hist(ax=ax, bins=25)
````

## Compare the ABV of wheat beers vs. IPAs : their IBUs were really different, but how about their alcohol percentage?

Wheat beers might include witbier, hefeweizen, American Pale Wheat Ale, and anything else you think is wheaty. IPAs probably have "IPA" in their name.

````python
df[df["Style"].isin(["Witbier","Hefeweizen","American Pale Wheat Ale"])].ABV.mean()  
````

````python
df[df.Style.str.contains('IPA.*', na=False)].ABV.mean() 
````

````python
# First I save the IPA histogram as ax
ax = df[df.Style.str.contains('IPA.*', na=False)].ABV.hist(bins=25)

# And then I include it in the wheat beer histogram
df[df["Style"].isin(["Witbier","Hefeweizen","American Pale Wheat Ale"])].ABV.hist(ax=ax, bins=25)
````

Looks like the ABV of the wheat beers is also lower than the ABV of the IPAs although the difference is not as big
