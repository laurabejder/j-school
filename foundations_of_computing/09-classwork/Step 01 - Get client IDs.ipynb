```python
import pandas as pd
```

```python
url = "https://cal-access.sos.ca.gov/Lobbying/Firms/list.aspx?letter=A"

tables = pd.read_html(url)
len(tables)
```

```python
tables[7]
```

```python
letters = ["A", "B", "C", "D", "E"]

dataframes = []
for letter in letters:
    url = f"https://cal-access.sos.ca.gov/Lobbying/Firms/list.aspx?letter={letter}"
    
    tables = pd.read_html(url)
    df = tables[7]
    
    # Add that page's info to the list
    dataframes.append(df)
print(dataframes)
    
```

```python
df = pd.concat(dataframes, ignore_index=True)
```

```python
df = df[df[0] != 'NAME']
df.columns = ['name', 'filer_id']
df.head()
```

```python
df.to_csv("firms.csv", index=False)
```
