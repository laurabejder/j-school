```python
# THE ABOVE ALSO INCLUDES SITUATIONS WHERE A PERSON WAS PUNCHED BY A WALL. 

# An easier and better way to do it would be: REGULAR EXPRESSIONS!

# if you want to put "... some stuff ..." 
#There are two ways to do it
    # at least one "some stuff"
    # 0 or more "some stuff"
    # * just means put whatever

punchers_df = df[df.Narrative_1.str.contains('PUNCH.*WALL', na=False)]
punchers_df.shape
# Apparently 2037 people punched walls. 
```
